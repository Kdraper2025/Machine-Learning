```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance
pd.set_option('display.max_columns', None)

credit_risk = pd.read_csv('credit_risk_dataset.csv')
credit_risk
```

# Remove target value

```python
credit_risk_subset = credit_risk[["person_age","person_income", "person_home_ownership","person_emp_length","loan_intent","loan_grade","loan_amnt", "loan_int_rate", "loan_percent_income","cb_person_default_on_file", "cb_person_cred_hist_length"]]
```

```python
credit_risk_subset
```

```python
target= credit_risk["loan_status"]
```

# Encode label categories

```python
le = LabelEncoder()
```

```python
credit_risk_subset["person_home_ownership"] = le.fit_transform(credit_risk_subset["person_home_ownership"])
credit_risk_subset[ "loan_intent"] = le.fit_transform(credit_risk_subset["loan_intent"])
credit_risk_subset["cb_person_default_on_file"] = le.fit_transform(credit_risk_subset["cb_person_default_on_file"])
```

```python
oe = OrdinalEncoder(categories=[["G", "E","F","D","C","B", "A" ]])
```

```python
credit_risk_subset["loan_grade"] = oe.fit_transform(credit_risk[["loan_grade"]])
```

# Check for nan values

```python
credit_risk_subset.isna().sum()
```

```python
mask =credit_risk_subset.isnull().any(axis=1)
print(credit_risk_subset[mask])
```

# Fill in values using KNNImputer

```python
from sklearn.impute import KNNImputer

knnimputer = KNNImputer(n_neighbors =3)
```

```python
credit_risk_subset.iloc[:,:] = knnimputer.fit_transform(credit_risk_subset)
```

```python
credit_risk_subset.isna().sum()
```

# Normalize the data

```python
norm_1 = tf.keras.layers.Normalization(axis =-1 )
norm_1.adapt(credit_risk_subset)
credit_risk_regularized = norm_1(credit_risk_subset)
```

```python
credit_risk_regularized
```

# Build your model

```python
model = Sequential(
    [
        Dense(200, activation = "relu"),
        Dense(50, activation = "relu"),
        Dense(30, activation = 'relu'),
        Dense(1, activation = "sigmoid")
        
    ]
)
```

```python
model.compile(optimizer = "adam" , loss = tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

# Split the data into training data, validation data and test data

```python
X_train, Xn, Y_train, Yn = train_test_split(credit_risk_regularized.numpy(), target, test_size =.50, random_state = 2)
X_cv, X_test, Y_cv, Y_test = train_test_split(Xn, Yn, test_size =.50, random_state = 2)
```

```python
print(f"X_train.shape {X_train.shape}, Y_train.shape {Y_train.shape}")
print(f"X_cv.shape {X_cv.shape}, Y_cv.shape {Y_cv.shape}")
print(f"X_test.shape {X_test.shape}, Y_test.shape {Y_test.shape}")
```

```python
X_train = tf.convert_to_tensor(X_train)
Y_train = tf.convert_to_tensor(Y_train)
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=6, mode = 'min')
```

```python
model.fit(X_train,Y_train, validation_data=(X_cv, Y_cv), epochs = 100, callbacks=[early_stop])
test_loss,test_acc=model.evaluate(X_test,Y_test)
```
